### 날씨 데이터 불러오기

```python
air= pd.read_csv('air.csv',encoding='ANSI')
air_2020= pd.read_csv('air_2020.csv',encoding='ANSI')
```

### 데이터 확인

* 2020년도 데이터로 나중에 test를 검증하기 위한 데이터 셋입니다.

* 년도가 너무 길어 시각화하기 힘들다는 점에서 2010년부터 살펴보겠습니다.

```python
air=air[(air['일시']>='2010-01-01')&(air['일시']<='2019-12-31')] 
```

```python
air['일시']=air['일시'].astype('str')
air['일시']=pd.to_datetime(air['일시'])
air.index=air['일시']
air.set_index('일시',inplace=True)
```

```python
air.plot()
```

* 아래 ARIMA모델을 작성할 때 사용되는 통계분석을 위한 패키지 statsmoel패키지 입니다.

```python
pip install statsmodels 
```

* AR(Autoregression)은 데이터의 과거를 정보로 사용한다는 개념=현재의 상태는 이전 상태에 따라 결정
* MA(Movingaverage)는 예측값과 실제값의 차이인데 이전항에서의 오차를 참고하여 결정
* I는 추세변동의 경향성

```python
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm

#(ar=2,차분=0, ma=2)파라미터로 arima모델을 학습
model=ARIMA(air.평균기온.values, order=(2,0,2))
model_fit=model.fit()
print(model_fit.summary())
```

```python
fig=model_fit.plot_predict()
residuals=pd.DataFrame(model_fit.resid)
residuals.plot()
```

```python
air_2020.index=air_2020['일시']
air_2020.set_index('일시',inplace=True)
```

```python
air_2020_h=air_2020.head(12)
```

```python
forecast_data=model_fit.forecast(steps=12) #학습데이터 셋으로 부터 10일 뒤를 예측

pred_y=forecast_data[0].tolist()
test_y=air_2020_h.평균기온.values #실제 10일의 미세먼지 데이터
```

```python
pred_y
```

```python
test_y
```

```python
plt.plot(pred_y,color='red')
plt.plot(test_y,color='blue')
```

```python
import itertools
p=d=q=range(0,5)
pdq=list(itertools.product(p,d,q))
pdq
```

```python
import warnings
warnings.filterwarnings('ignore')

for param in pdq:
    try:
        model_arima=ARIMA(air.평균기온.values, order=param)
        model_arima_fit=model_arima.fit()
        print(param,model_arima_fit.aic)
    except:
        continue
```

### 예측성능 평가

```python
from sklearn.metrics import mean_squared_error,r2_score
from math import sqrt

rmse=sqrt(mean_squared_error(pred_y,test_y))
print(rmse)
```

# [Facebook Prophet활용하기]

```python
from fbprophet import Prophet
```

```python
air.columns=['ds','y']
air_2020.columns=['ds','y']
```

```python
air
```

```python
air=air[(air['ds']>='2010-01-01')&(air['ds']<='2019-12-31')] 
```

```python
air
```

```python
prophet = Prophet(seasonality_mode = 'multiplicative',
                 yearly_seasonality=True, 
                 weekly_seasonality=True,
                 daily_seasonality=True,
                 changepoint_prior_scale=0.5)
prophet.fit(air)
```

```python
future_data = prophet.make_future_dataframe(periods = 15, freq = 'm')
forecast_data = prophet.predict(future_data)
forecast_data[['ds','yhat', 'yhat_lower', 'yhat_upper']].tail(15)
```

```python
fig1=prophet.plot(forecast_data)
```

```python
pred_y=forecast_data.yhat.values[121:132]
test_y=air_2020.y.values[0:11]
```

```python
pred_y
```

```python
test_y
```

```python
plt.plot(pred_y,color='red')
plt.plot(test_y,color='blue')
```

```python
from sklearn.metrics import mean_squared_error,r2_score
from math import sqrt
rmse=sqrt(mean_squared_error(pred_y,test_y))
print(rmse)
```
