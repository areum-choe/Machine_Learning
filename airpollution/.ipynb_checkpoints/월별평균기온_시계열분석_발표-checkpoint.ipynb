### 아래 statsmoel패키지는 ARIMA모델을 작성할 때 사용되는 통계분석을 위한 패키지입니다.

```python
pip install statsmodels 
```

* AR(Autoregression)은 데이터의 과거를 정보로 사용한다는 개념=현재의 상태는 이전 상태에 따라 결정
* MA(Movingaverage)는 예측값과 실제값의 차이인데 이전항에서의 오차를 참고하여 결정
* I는 추세변동의 경향성

```python
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm

#(ar=2,차분=0, ma=2)파라미터로 arima모델을 학습
model=ARIMA(air.평균기온.values, order=(2,0,2))
model_fit=model.fit()
print(model_fit.summary())
```

```python
fig=model_fit.plot_predict()
residuals=pd.DataFrame(model_fit.resid)
residuals.plot()
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(air)
plot_pacf(air)
plt.show()
```

```python
diff_1=air.diff(periods=1).iloc[1:]
diff_1.plot()
plot_acf(diff_1)
plot_pacf(diff_1)
plt.show()
```

#### 2020년도의 예측이 맞는지 확인

```python
air_2020.index=air_2020['일시']
air_2020.set_index('일시',inplace=True)
```

```python
air_2020_h=air_2020.head(12)
```

```python
forecast_data=model_fit.forecast(steps=12) 

pred_y=forecast_data[0].tolist()
test_y=air_2020_h.평균기온.values 
```

```python
pred_y
```

```python
test_y
```

```python
plt.plot(pred_y,color='red')
plt.plot(test_y,color='blue')
```

### 위 ARIMA의 p,d,q를 설정하기 위한 방법

```python
import itertools
p=d=q=range(0,5)
pdq=list(itertools.product(p,d,q))
pdq
```

```python
import warnings
warnings.filterwarnings('ignore')

for param in pdq:
    try:
        model_arima=ARIMA(air.평균기온.values, order=param)
        model_arima_fit=model_arima.fit()
        print(param,model_arima_fit.aic)
    except:
        continue
```

* 위에서 가장 낮은 숫자를 찾아 대입

#### 예측성능 평가

```python
from sklearn.metrics import mean_squared_error,r2_score
from math import sqrt

rmse=sqrt(mean_squared_error(pred_y,test_y))
print(rmse)
```

## [Facebook Prophet활용하기]

```python
from fbprophet import Prophet
```

```python
air.columns=['ds','y']
air_2020.columns=['ds','y']
```

```python
air=air[(air['ds']>='2010-01-01')&(air['ds']<='2019-12-31')] 
```

```python
prophet = Prophet(seasonality_mode = 'multiplicative',
                 yearly_seasonality=True, 
                 weekly_seasonality=True,
                 daily_seasonality=True,
                 changepoint_prior_scale=0.5)
prophet.fit(air)
```

```python
future_data = prophet.make_future_dataframe(periods = 370, freq = 'd')
forecast_data = prophet.predict(future_data)
a=forecast_data[['ds','yhat', 'yhat_lower', 'yhat_upper']].tail(370)
```

```python
fig1=prophet.plot(forecast_data)
```

```python
a['day']=a['ds'].dt.day
```

```python
a=a[a['day']==1] 
```

```python
a
```

```python
a.reset_index(drop=True)
```

```python
pred_y=a.yhat.values[0:11]
test_y=air_2020.y.values[0:11]
```

```python
pred_y
```

```python
test_y
```

```python
plt.plot(pred_y,color='red')
plt.plot(test_y,color='blue')
```

```python
from sklearn.metrics import mean_squared_error,r2_score
from math import sqrt
rmse=sqrt(mean_squared_error(pred_y,test_y))
print(rmse)
```
