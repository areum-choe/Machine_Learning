```python
import pandas as pd
excel_url = 'C:/Users/areum/Desktop/POC시연자료/마이코드몰_주문일기준_2021년_1년자료.xlsx'
# 첫번째 시트를 읽어옵니다.
df1 = pd.read_excel(excel_url, sheet_name = 0)
# 두번째 시트를 읽어옵니다.
df2 = pd.read_excel(excel_url, sheet_name = 1)
```

## 회원 주소를 결합하기 위한 과정

```python
a=pd.merge(df1, df2, left_on = '회원 아이디', right_on = '아이디', how = 'inner')
```

```python
a=a.drop_duplicates(['회원 아이디'])
```

```python
a=a[['회원명','회원 아이디', '성별','생일','수취인 주소']]
```

```python
a["생일"]=a["생일"].astype('str')
```

```python
a.info()
```

```python
a['생일']= a['생일'].str.slice(start=0, stop=4)
```

```python
a
```

```python
name_split = a["수취인 주소"].str.split(" ")
```

```python

a["수취인 주소"] = name_split.str.get(0)
```

```python
a
```

```python
a=a.dropna(axis=0)
```

```python
a=a.reset_index(drop=True)
```

```python
a
```

```python
a.to_excel(excel_writer='C:/Users/areum/Desktop/POC시연자료/마이코드몰sql.xlsx')
```
